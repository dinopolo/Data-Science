# **Uma análise da COVID-19 (Coronavírus) no Mundo e no Brasil**

A COVID-19 é uma doença causada pelo coronavírus da síndrome respiratória aguda 2 (SARS-CoV-2). Foi identificada pela primeira em Wuhan, província de Hubei na China, em dezembro de 2019 e, em março de 2020, a Organização Mundial de Saúde (OMS - em inglês: *World Health Organization - WHO*) declarou o surto de COVID-19 como pandemia.

Os sintomas mais comuns da doença são:
- Febre;
- Tosse;
- Cansaço.

Nos estágios iniciais podem ocorrer:
- Perda de olfato e paladar.

Ainda, outros sintomas que podem ocorrer:
- Dificuldade de respirar;
- Dores musculares;
- Calafrios;
- Dor de garganta;
- Coriza nasal;
- Dor de cabeça;
- Dor no peito.

Também, sintomas menos comuns podem incluir:
- Erupção cutânea;
- Náusea;
- Vômito;
- Diarreia.

O coronavírus se espalha mais facilmente de pessoa a pessoa, principalmente, quando estão próximas (abaixo de 2 metros). Além disso, o vírus se propaga por gotículas respiratórias, quando uma pessoa infectada tosse, espirra ou fala; e essas gotículas podem ser inaladas ou cair na boca ou nariz de uma pessoa próxima. Também, pode se propagar quando alguém toca uma superfície com o coronavírus e, então, toca o seu nariz, boca ou olhos, no entanto, essa não é considerada uma forma comum de propagação.

Além disso, aproximadamente 80% dos casos confirmados são leves ou assintomáticos e a maioria se recupera sem problemas. No entnanto, em torno de 15% são infecções graves que precisam de oxigênio e 5% são casos graves que necessitam de ventilação mecânica. Ainda, os casos mais graves podem evoluir para pneumonia grave com insuficiência respiratória grave, sepse, falência múltipla de órgãos e morte.

Sendo assim, com objetivo de informar mais as pessoas, realizarei uma análise utilizando dados públicos sobre a COVID-19.

**Comparativo entre o número de casos confirmados e as mortes pela COVID-19**

Faremos um gráfico de linhas comparando número de casos confirmados e as mortes pela COVID-19, e assim podemos avaliar se esses dois fatores, de fato, apresentam um comportamento semelhante.

Avaliando a Progressão da taxa de mortes pela COVID-19, percebemos que:
- O seu pico (em torno de 7%) foi atingindo aproximadamente no 125º dia da Pandemia, e depois começou a cair, isso pode ter ocorrido devido aos profissionais da área de saúde terem conhecido mais sobre a doença e assim desenvolvido melhores meios de combatê-la.
- Nos dias iniciais da pandemia, a taxa de mortes oscilou consideravelmente, isso ocorre, principalmente, por se tratar de uma doença nova e as pessoas não entenderam exatamente como funciona o seu mecanismo de progressão em pessoas infectadas, além disso, nesses dias iniciais, também não fica claro como o corpo humano combate a doença.
- É importante notar também, que no dia mais recente da coleta dos dados, a taxa de mortes mostra tendência de queda, o que é algo positivo e, possivelmente, os tratamentos estão sendo mais eficazes para uma recuperação dos pacientes.
